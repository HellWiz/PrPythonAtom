```python
class HashMap:
    
    class Entry:
        def __init__(self, key, value):
            #TODO store key and value as attributes
            self.key = key
            self.value = value
            
        def get_key(self):
            #TODO return key
            return self.key
            
        def get_value(self):
            #TODO return value
            return self.value
            
        def __eq__(self, other):
            #TODO реализовать функцию сравнения
            #Пояснения: self - текущий объект entry, other - какой-то другой объект
            #Какое условие на равенство объектов в ассоциативном массиве?
            #Если объекты равны, то метод должен возвращать True, иначе False
            return self.key == other.key
    
    def __init__(self, bucket_num = 64):
        #TODO Создать массив для бакетов
        self.capacity = bucket_num
        self.size = 0
        self.inside = []
        for i in range(bucket_num):
            self.inside.append([])        

    def get(self, key, default_value = None):
        #TODO метод get, возвращающий значение, если оно присутствует, иначе default_value
        for i in self.inside[self._get_index(self._get_hash(key))]:
            if i.key == key:
                return i.value
        return default_value         

    def put(self, key, value):
        #TODO метод put, кладет значение по ключу, в случае, если ключ уже присутствует он его заменяет
        for i in self.inside[self._get_index(self._get_hash(key))]:
            if i.key == key:
                i.value = value
                return
        item = self.Entry(key, value)
        self.inside[self._get_index(self._get_hash(key))].append(item)
        self.size += 1

    def __len__(self):
        #TODO Возвращает количество Entry в массиве
        return self.size

    def _get_hash(self, key):
        #TODO Вернуть хеш от ключа, по которому он кладется в бакет
        return hash(key)

    def _get_index(self, hash_value):
        #TODO По значению хеша вернуть индекс элемента в массиве
        return hash_value%self.capacity
        
    class Val_iter:
        def __iter__(self):
            return self
        
        def __init__(self,obj):
            self.obj = obj
            self.limit = obj.size
            self.count = 0
            self.cur_pos = 0
            self.capacity = obj.capacity
            for i in range(self.capacity):
                if len(obj.inside[i]) > 0:
                    self.cur_arr = i
                    return
                    
        def __next__(self):
            if self.count < self.limit:
                self.count += 1
                prev_pos = self.cur_pos
                if self.cur_pos == len(self.obj.inside[self.cur_arr]):
                    self.cur_arr += 1
                    while len(self.obj.inside[self.cur_arr]) == 0:
                        self.cur_arr += 1
                    self.cur_pos = 0
                    prev_pos = 0
                self.cur_pos += 1
                return self.obj.inside[self.cur_arr][prev_pos].value
            else:
                raise StopIteration
                
    class Key_iter:
        def __iter__(self):
            return self
        
        def __init__(self,obj):
            self.obj = obj
            self.limit = obj.size
            self.count = 0
            self.cur_pos = 0
            self.capacity = obj.capacity
            for i in range(self.capacity):
                if len(obj.inside[i]) > 0:
                    self.cur_arr = i
                    return
                    
        def __next__(self):
            if self.count < self.limit:
                self.count += 1
                prev_pos = self.cur_pos
                if self.cur_pos == len(self.obj.inside[self.cur_arr]):
                    self.cur_arr += 1
                    while len(self.obj.inside[self.cur_arr]) == 0:
                        self.cur_arr += 1
                    self.cur_pos = 0
                    prev_pos = 0
                self.cur_pos += 1
                return self.obj.inside[self.cur_arr][prev_pos].key
            else:
                raise StopIteration
                
    class Item_iter:
        def __iter__(self):
            return self
        
        def __init__(self,obj):
            self.obj = obj
            self.limit = obj.size
            self.count = 0
            self.cur_pos = 0
            self.capacity = obj.capacity
            for i in range(self.capacity):
                if len(obj.inside[i]) > 0:
                    self.cur_arr = i
                    return
                    
        def __next__(self):
            if self.count < self.limit:
                self.count += 1
                prev_pos = self.cur_pos
                if self.cur_pos == len(self.obj.inside[self.cur_arr]):
                    self.cur_arr += 1
                    while len(self.obj.inside[self.cur_arr]) == 0:
                        self.cur_arr += 1
                    self.cur_pos = 0
                    prev_pos = 0
                self.cur_pos += 1
                res = tuple()
                res = (self.obj.inside[self.cur_arr][prev_pos].key,
                             self.obj.inside[self.cur_arr][prev_pos].value)
                return res
            else:
                raise StopIteration
                
    def values(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор значений (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        return self.Val_iter(self)
        
    def keys(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор ключей (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        return self.Key_iter(self)
        
    def items(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор пар ключ и значение (tuples)
        #(на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        return self.Item_iter(self)
        
    def __str__(self):
        #TODO Метод, вызываемый при вызове print, пофантазируйте, что можно написать
        res = ''
        for i in self.inside:
            for j in i:
                res+='({}, {})'.format(j.key, j.value)
        return res
```

```python
class HashSet(HashMap):
   
    def get(self, key, default_value=None):
        #TODO достаточно переопределить данный метод
        for i in self.inside[self._get_index(self._get_hash(key))]:
            if i.key == key:
                return True
        return default_value     

    def put(self, key, value=True):
        #TODO метод put, нужно переопределить данный метод
        for i in self.inside[self._get_index(self._get_hash(key))]:
            if i.key == key:
                return
        item = self.Entry(key, value)
        self.inside[self._get_index(self._get_hash(key))].append(item)
        self.size += 1

    def __len__(self):
        #TODO Возвращает количество Entry в массиве
        return self.size

    def values(self):
        #TODO Задание со звездочкой!
        #Если делали, то нужно бы переопределить
        #Должен возвращать генератор или итератор значений (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        return self.Key_iter(self)
    #      наверное еще имело смысл переопределить __str__
```

```python
import random

def test_hashmap_04():
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ()), ({"s":"v"}, {"v":"s"})]
    for k, v in entries:
        entry = HashMap.Entry(k, v)
        assert entry.get_key() == k
        assert entry.get_value() == v
    print("Test 4 part 01 passed")
        
    for i in range(len(entries)):
        entry_one = HashMap.Entry(entries[i][0], entries[i][1])
        for _ in range(10):
            j = random.randint(0, len(entries)-1)
            p = random.randint(0, len(entries)-1)
            entry_two = HashMap.Entry(entries[j][0], entries[p][1])
            if j == i:
                assert entry_one == entry_two
            else:
                assert entry_one != entry_two
    print("Test 4 part 02 passed")
    
def test_hashmap_05():
    hashmap = HashMap(10)
    assert sum(isinstance(v, list) for k, v in hashmap.__dict__.items()) == 1
    print("Test 5 part 01 passed")
    
    inner_list_name = [k for k, v in hashmap.__dict__.items() if isinstance(v, list)][0]
    for i in range(10):
        assert len(HashMap(i).__dict__[inner_list_name]) == i
    print("Test 5 part 02 passed")
    
def test_hashmap_06():
    hashmap = HashMap(10)
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ())]
    for k, v in entries:
        hashmap.put(k, v)
    for k, v in entries:
        assert hashmap.get(k) == v
    print("Test 6 part 01 passed")
    
    for _ in range(100):
        i = random.randint(0, len(entries)-1)
        j = random.randint(0, len(entries)-1)
        hashmap.put(i, j)
        assert hashmap.get(i) == j
        
    assert hashmap.get("nexit", "default") =="default"
    
    print("Test 6 part 02 passed")
```

```python
test_hashmap_04()
test_hashmap_05()
test_hashmap_06()
```
